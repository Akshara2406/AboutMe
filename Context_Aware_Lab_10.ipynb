```python
class Refrigerator:
  def __init__(self,Level_R,Temp_R,Door_R,Weight_R,Power_R,GPSLA_R,GPSLO_R):
    self.GPSLA_R=GPSLA_R
    self.GPSLO_R=GPSLO_R
    self.Temp_R=Temp_R
    self.Door_R=Door_R
    self.Weight_R=Weight_R
    self.Power_R=Power_R
    self.Level_R=Level_R
  def display(self):
    if self.GPSLA_R>0 and self.GPSLO_R>0:
      print()
      print("Appliance is Refrigerator")
      print("Power: ",self.Power_R)
      if (self.Power_R > 3.0):
        print("Refrigerator is ON")
        print("Weight of Coolant: ",self.Weight_R)
        if self.Weight_R>33:
          print("Coolant Sufficient")
          print("Temperature: ",self.Temp_R)
          if self.Temp_R>15:
            print("Cooling begins")
          else:
            print("Cooling Sufficient")
          print("Door Angle: ",self.Door_R)
          if self.Door_R!=0:
            print("Door Open,Buzzer ON")
          else:
            print("Door Close,Buzzer OFF")   
          print("Waste Water Level: ",self.Level_R)
          if self.Level_R>1:
            print("Water outlet OPEN")
          else:
            print("Water outlet CLOSE")
        else:
          print("Coolant Insufficient")
      else:
        print("Refrigerator is OFF")
  

class AirConditioner:
  def __init__(self,AirQua_A,Door_A,Temp_A,Weight_A,Power_A,GPSLA_A,GPSLO_A,PIR_2):
    self.GPSLA_A=GPSLA_A
    self.GPSLO_A=GPSLO_A
    self.Temp_A=Temp_A
    self.Weight_A=Weight_A
    self.Power_A=Power_A
    self.AirQua_A=AirQua_A
    self.PIR_2=PIR_2
    self.Door_A=Door_A
  def display(self):
    if self.GPSLA_A>0 and self.GPSLO_A>0:
      print()
      print("Appliance is Air Conditioner")
      print("Power: ",self.Power_A)
      if self.PIR_2 == 1 and self.Power_A > 1:
        print("Person Detected,AC is ON")
        if self.Door_A == 0:
          print("Room Door Closed")
          print("Weight of Coolant Gas : ",self.Weight_A)
          if self.Weight_A>=700:
            print("Coolant Sufficient")
            print("Temperature: ",self.Temp_A)
            if self.Temp_A>35:
              print("Cooling begins")
            else:
              print("Cooling Sufficient")
            print("Air Quality: ",self.AirQua_A)
            if self.AirQua_A<750 :
              print("Clean Air")
            else:
              print("Unclean Air")
          else:
            print("Coolant Insufficient")
        else:
          print("Room Door Open")
      else:
        print("Person not Detected,AC is OFF")
      
      
class Geaser:
  def __init__(self,Level_G,Temp_G,Power_G,GPSLA_G,GPSLO_G,PIR_3):
    self.GPSLA_G=GPSLA_G
    self.GPSLO_G=GPSLO_G
    self.Temp_G=Temp_G
    self.Level_G=Level_G
    self.Power_G=Power_G
    self.PIR_3=PIR_3
  def display(self):
    if self.GPSLA_G>0 and self.GPSLO_G>0:
      print()
      print("Appliance is Geaser")
      print("Power: ",self.Power_G)
      if self.PIR_3 == 1 or self.Power_G > 1:
        print("Geyser is ON")
        print("Level of water : ",self.Level_G)
        if self.Level_G>=5:
          print("Water Sufficient")
          print("Temperature: ",self.Temp_G)
          if self.Temp_G<40:
            print("Heating begins")
          else:
            print("Heating Sufficient")
        else:
          print("Water Insufficient")
      else:
        print("Geyser is OFF")

class Microwave:
  def __init__(self,Temp_M,Door_M,Weight_M,Power_M,GPSLA_M,GPSLO_M,PIR_1):
    self.GPSLA_M=GPSLA_M
    self.GPSLO_M=GPSLO_M
    self.Temp_M=Temp_M
    self.Door_M=Door_M
    self.Weight_M=Weight_M
    self.Power_M=Power_M
    self.PIR_1=PIR_1
  def display(self):
    if self.GPSLA_M>0 and self.GPSLO_M>0:
      print()
      print("Appliance is Microwave")
      print("Power: ",self.Power_M)
      if self.PIR_1 == 1 or self.Power_M > 1:
        print("Microwave is ON")
        print("Door Angle: ",self.Door_M)
        if self.Door_M==0:
          print("Door Close,Buzzer OFF")
          print("Weight of Food: ",self.Weight_M)
          if self.Weight_M>0: 
            print("Food Inside,Turn ON Microwave")
          else:
            print("No food Inside,PREHEAT MODE")
          print("Temperature: ",self.Temp_M)
          if self.Temp_M>50:
            print("Fan is turned ON")
          else:
            print("Fan is turned OFF")
        else:
          print("Door Open,Buzzer ON")
      else:
        print("Microwave is OFF")


class WashingMachine:
  def __init__(self,Level_W,Door_W,Weight_W,Power_W,GPSLA_W,GPSLO_W):
    self.GPSLA_W=GPSLA_W
    self.GPSLO_W=GPSLO_W
    self.Door_W=Door_W
    self.Level_W=Level_W
    self.Weight_W=Weight_W
    self.Power_W=Power_W
  def display(self):
    if self.GPSLA_W>0 and self.GPSLO_W>0:
      print()
      print("Appliance is Washing Machine")
      print("Power: ",self.Power_W)
      if self.Power_W > 1:
        print("Washing Maching is ON")
        print("Door Angle: ",self.Door_W)
        print("Weight of clothes: ",self.Weight_W)
        if self.Door_W==0 and self.Weight_W>300:
          print("Door Close and enough clothes,Buzzer OFF")
          print("Level of Water: ",self.Level_W)
          if self.Level_W>25:
            print("Water Sufficient")
          else:
            print("Water Insufficient")
        else:
          print("Door Open/No Clothes,Buzzer ON")
      else:
        print("Washing Machince is OFF")




```

```python
from IPython.display import Javascript
display(Javascript('''google.colab.output.setIframeHeight(0, true, {maxHeight: 5000})'''))
class Context:
  def __init__(self,P):
    self.Refrigerator_=Refrigerator(energy['Level_R'][P],energy['Temp_R'][P],energy['Door_R'][P],energy['Weight_R'][P],energy['Power_R'][P],energy['GPSLA_R'][P],energy['GPSLO_R'][P])
    self.AirConditioner_=AirConditioner(energy['AirQua_A'][P],energy['Door_A'][P],energy['Temp_A'][P],energy['Weight_A'][P],energy['Power_A'][P],energy['GPSLA_A'][P],energy['GPSLO_A'][P],energy['PIR_2'][P])
    self.Geaser_=Geaser(energy['Level_G'][P],energy['Temp_G'][P],energy['Power_G'][P],energy['GPSLA_G'][P],energy['GPSLO_G'][P],energy['PIR_3'][P])
    self.Microwave_=Microwave(energy['Temp_M'][P], energy['Door_M'][P], energy['Weight_M'][P], energy['Power_M'][P], energy['GPSLA_M'][P],energy['GPSLO_M'][P],energy['PIR_1'][P])
    self.WashingMachine_=WashingMachine(energy['Level_W'][P],energy['Door_W'][P],energy['Weight_W'][P],energy['Power_W'][P],energy['GPSLA_W'][P],energy['GPSLO_W'][P])


  def display(self):
    self.Refrigerator_.display()
    self.AirConditioner_.display()
    self.Geaser_.display()
    self.Microwave_.display()
    self.WashingMachine_.display()

if __name__=="__main__":
  counter=1
  contextList=[]
  for i in range(20):
    context = Context(i)
    contextList.append(context)
  for context in contextList:
    print("------------------Context Behaviour---------------------")
    print()
    context.display()
    counter=counter+1
    print()
    print()
```

### ***CASE BASED:***

**INITIAL RULE:** IF **Refrigerator:** OFF/ON,**Coolant:** InSufficient 
**---->** 
**Temperature:** No changes,**Door:** No Changes, **Outlet:** No changes

**NEW RULE BASED DERIVED FROM COOLANT SUFFICIENT:**  IF **Refrigerator:** OFF/ON,**Coolant:** InSufficient 
**---->**
 **Temperature:** LESS,**Door:** CLOSE, **Outlet:** CLOSE


So, **Hydrogen addition:**YES, **ALARM:** OFF, **OUTLET:** CLOSE

In Refrigerator if the coolant is not enough then coolant addition happens at that time door should be closed and temp should be less so that coolant would be required less based on past analysis the temperature needed for better coolant retainity is used to predict the temperature needed to be maintained.
